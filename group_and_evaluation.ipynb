```python
categories = {
    "ethnicity": ["ethnicity_Caucasoid", "ethnicity_Mongoloid", "ethnicity_Negroid", "ethnicity_Unsure"],
    "age": ["age_Adult", "age_Old", "age_Young"],
    "gender": ["gender_Female", "gender_Male"],
    "hair_length": ["hair_length_Bald", "hair_length_Long", "hair_length_Short"],
    "upper_body_length": ["upper_body_length_Long", "upper_body_length_Short"],
    "upper_body_color": ["upper_body_color_Black", "upper_body_color_Blue", "upper_body_color_Brown", "upper_body_color_Green", "upper_body_color_Grey", "upper_body_color_Orange", "upper_body_color_Other", "upper_body_color_Pink", "upper_body_color_Purple", "upper_body_color_Red", "upper_body_color_White", "upper_body_color_Yellow"],
    "upper_body_type": ["upper_body_type_Blouse", "upper_body_type_Dress", "upper_body_type_Jacket", "upper_body_type_Other", "upper_body_type_Polo", "upper_body_type_Shirt", "upper_body_type_Tanktop", "upper_body_type_Tshirt"],
    "lower_body_length": ["lower_body_length_Long", "lower_body_length_Short"],
    "lower_body_color": ["lower_body_color_Black", "lower_body_color_Blue", "lower_body_color_Brown", "lower_body_color_Green", "lower_body_color_Grey", "lower_body_color_Orange", "lower_body_color_Other", "lower_body_color_Pink", "lower_body_color_Purple", "lower_body_color_Red", "lower_body_color_White", "lower_body_color_Yellow"],
    "lower_body_type": ["lower_body_type_Skirt&Dress", "lower_body_type_Trousers&Shorts"],
    "footwear": ["footwear_Sandals", "footwear_Shoes"],
    "backpack": ["backpack_Yes"],
    "bag": ["bag_Yes"],
    "glasses": ["glasses_No", "glasses_Normal", "glasses_Sun"],
    "hat": ["hat_Yes"],
    "mask": ["mask_Yes"]
}
```

```python
# Function to get the prediction for each category
def get_predictions(row, categories):
    predictions = {}
    for category, columns in categories.items():
        if category in ["backpack", "bag", "hat", "mask"]:
            # Apply thresholding for "Yes" classes
            for col in columns:
                if row[col] > 0.5:
                    predictions[category] = col.split('_')[1]
                    break
            else:
                predictions[category] = "No"
        else:
            # Find the column with the highest probability for the current category
            max_col = row[columns].idxmax()
            if category.endswith("length") or category.endswith("type") or category.endswith("color"):
                predictions[category] = max_col.split('_')[-1]
            else:
                predictions[category] = max_col.split('_', 1)[1]
    return predictions
```

```python
# Function to get the prediction for each category
def get_predictions_encode(row, categories):
    predictions = {}
    for category, columns in categories.items():
        if category in ["backpack", "bag", "hat", "mask"]:
            # Apply thresholding for "Yes" classes
            for col in columns:
                if row[col] > 0.5:
                    predictions[category] = 1
                    break
            else:
                predictions[category] = 0
        else:
            # Initialize prediction for current category
            max_col = row[columns].idxmax()
            for col in columns:
                if col == max_col:
                    predictions[col] = 1
                else:
                    predictions[col] = 0
    return predictions
# Apply the function to each row in the DataFrame
encode_predicted_df = prob_df.apply(lambda row: pd.Series(get_predictions_encode(row, categories)), axis=1)

encode_predicted_df.rename(columns={
    'backpack': 'backpack_Yes',
    'bag': 'bag_Yes',
    'hat': 'hat_Yes',
    'mask': 'mask_Yes'
}, inplace=True)
```

```python
# Apply the function to each row in the DataFrame
predicted_df = prob_df.apply(lambda row: pd.Series(get_predictions(row, categories)), axis=1)
# Save the averages to a new CSV file
predicted_df.to_csv(f'real_prediction_{model_name}.csv')
predicted_df
```

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score

ground_truth_df = pd.read_csv("/home/noeyiue/cpe/ICSEC/annotation/test_data.csv", index_col="image_name")
ground_truth_df = ground_truth_df.astype(bool)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score

# Assuming `categories` and `encode_predicted_df` are defined as per your previous context

# Initialize a dictionary to store performance metrics
performance_metrics = {
    "category": [],
    "TP": [],
    "TN": [],
    "FP": [],
    "FN": [],
    "Precision": [],
    "Recall": [],
    "F1-score": []
}

# Calculate metrics for each category
for main_category, sub_categories in categories.items():
    for category in sub_categories:
        # Convert predictions to boolean based on threshold (assuming 0.5 for binary classification)
        predicted = encode_predicted_df[category].astype(bool)
        actual = ground_truth_df[category]
        
        # Calculate TP, TN, FP, FN
        TP = ((predicted == True) & (actual == True)).sum()
        TN = ((predicted == False) & (actual == False)).sum()
        FP = ((predicted == True) & (actual == False)).sum()
        FN = ((predicted == False) & (actual == True)).sum()
        
        # Calculate precision, recall, and F1-score
        precision = precision_score(actual, predicted, zero_division=0)
        recall = recall_score(actual, predicted)
        f1 = f1_score(actual, predicted)
        
        # Append results to the performance_metrics dictionary
        performance_metrics["category"].append(category)
        performance_metrics["TP"].append(TP)
        performance_metrics["TN"].append(TN)
        performance_metrics["FP"].append(FP)
        performance_metrics["FN"].append(FN)
        performance_metrics["Precision"].append(precision)
        performance_metrics["Recall"].append(recall)
        performance_metrics["F1-score"].append(f1)

# Convert performance_metrics dictionary to DataFrame
performance_metrics_df = pd.DataFrame(performance_metrics)

# Save performance_metrics_df to a CSV file
performance_metrics_df.to_csv('performance_metrics.csv', index=False)
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Create a directory to save confusion matrix images if it doesn't exist
save_dir = f'./confusion_matrix_{model_name}/'
os.makedirs(save_dir, exist_ok=True)

# Iterate over main categories, excluding specific ones
excluded_categories = ['backpack', 'bag', 'hat', 'mask']
for main_category, sub_categories in categories.items():
    if main_category not in excluded_categories:
        # Process regular categories
        cat_predicted = encode_predicted_df[sub_categories]
        cat_predicted_values = cat_predicted.values.tolist()
        cat_gt = ground_truth_df[sub_categories]
        cat_gt_values = cat_gt.values.tolist()
        
        binary_cat_predicted = np.array([np.argmax(item) for item in cat_predicted_values])
        binary_cat_gt = np.array([np.argmax(item) for item in cat_gt_values])
        
        cm = confusion_matrix(binary_cat_gt, binary_cat_predicted)
        
        # Plot confusion matrix
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, cmap='Blues', fmt='d', xticklabels=sub_categories, yticklabels=sub_categories)
        plt.title(f'Confusion Matrix for {main_category}')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        
        # Save the plot as an image
        save_path = os.path.join(save_dir, f'{main_category}_cm.jpg')
        plt.savefig(save_path)
        plt.close()  # Close the plot to free memory
        
    else:
        # Process 'Yes/No' categories
        binary_cat_predicted = encode_predicted_df[sub_categories].values.tolist()
        binary_cat_gt = ground_truth_df[sub_categories].values.tolist()
        
        cm = confusion_matrix(binary_cat_gt, binary_cat_predicted)
        
        # Plot confusion matrix
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, cmap='Blues', fmt='d', xticklabels=["Yes", "No"], yticklabels=["Yes", "No"])
        plt.title(f'Confusion Matrix for {main_category}')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        
        # Save the plot as an image
        save_path = os.path.join(save_dir, f'{main_category}_cm.jpg')
        plt.savefig(save_path)
        plt.close()  # Close the plot to free memory
```
